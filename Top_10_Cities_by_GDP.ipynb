# Top 10 Cities by GDP
This notebook identifies and visualizes the top 10 metropolitan areas/cities with the highest GDP.

```python

import pandas as pd
import matplotlib.pyplot as plt

file_path = 'data/cities_by_gdp.csv'
data = pd.read_csv(file_path)

data['Official est. GDP(billion US$)'] = pd.to_numeric(data['Official est. GDP(billion US$)'].str.replace(',', ''))
top_10_gdp = data.sort_values(by='Official est. GDP(billion US$)', ascending=False).head(10)

# Plotting the top 10 cities by GDP
plt.figure(figsize=(12, 8))
plt.bar(top_10_gdp['Metropolitian Area/City'], top_10_gdp['Official est. GDP(billion US$)'])
plt.xlabel('City')
plt.ylabel('GDP (billion US$)')
plt.title('Top 10 Cities by GDP')
plt.xticks(rotation=45, ha='right')
plt.show()
```
