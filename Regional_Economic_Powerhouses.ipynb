# Regional Economic Powerhouses
This notebook identifies and visualizes the economic powerhouses within each region.

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'data/cities_by_gdp.csv'
data = pd.read_csv(file_path)

data['Official est. GDP(billion US$)'] = pd.to_numeric(data['Official est. GDP(billion US$)'].str.replace(',', ''))

# Plotting the economic powerhouses within each region
regions = data['Country/Region'].unique()
fig, axs = plt.subplots(len(regions), 1, figsize=(14, len(regions) * 4), sharex=True)

for i, region in enumerate(regions):
    region_data = data[data['Country/Region'] == region]
    top_city = region_data.sort_values(by='Official est. GDP(billion US$)', ascending=False).head(1)
    axs[i].bar(top_city['Metropolitian Area/City'], top_city['Official est. GDP(billion US$)'])
    axs[i].set_title(f'{region} Economic Powerhouse')
    axs[i].set_ylabel('GDP (billion US$)')
    axs[i].set_xticks(range(len(top_city['Metropolitian Area/City'])))
    axs[i].set_xticklabels(top_city['Metropolitian Area/City'], rotation=45, ha='right')

plt.xlabel('City')
plt.tight_layout()
plt.show()
```
