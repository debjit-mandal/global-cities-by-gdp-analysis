# GDP vs. Population Correlation
This notebook analyzes the relationship between GDP and population of metropolitan areas/cities.

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'data/cities_by_gdp.csv'
data = pd.read_csv(file_path)

data['Official est. GDP(billion US$)'] = pd.to_numeric(data['Official est. GDP(billion US$)'].str.replace(',', ''))
data['Metropolitian Population'] = pd.to_numeric(data['Metropolitian Population'].str.replace(',', '').str.replace('', '').str.strip())

# Scatter plot of GDP vs. Population
plt.figure(figsize=(12, 8))
plt.scatter(data['Metropolitian Population'], data['Official est. GDP(billion US$)'])
plt.xlabel('Population')
plt.ylabel('GDP (billion US$)')
plt.title('GDP vs. Population')
plt.xscale('log')
plt.yscale('log')
plt.show()

# Correlation coefficient
correlation = data['Official est. GDP(billion US$)'].corr(data['Metropolitian Population'])
print(f'Correlation coefficient between GDP and Population: {correlation}')
```
