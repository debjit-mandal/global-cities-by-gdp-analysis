# Comparison of Mega Cities
This notebook compares the GDP and population of mega cities with populations over 10 million.

```python

import pandas as pd
import matplotlib.pyplot as plt

file_path = 'data/cities_by_gdp.csv'
data = pd.read_csv(file_path)

data['Official est. GDP(billion US$)'] = pd.to_numeric(data['Official est. GDP(billion US$)'].str.replace(',', ''))
data['Metropolitian Population'] = pd.to_numeric(data['Metropolitian Population'].str.replace(',', '').str.replace(' ', '').str.strip())

# Mega cities (population > 10 million)
mega_cities = data[data['Metropolitian Population'] > 10000000]

# Plotting the comparison of mega cities
fig, ax1 = plt.subplots(figsize=(12, 8))

ax2 = ax1.twinx()
ax1.bar(mega_cities['Metropolitian Area/City'], mega_cities['Official est. GDP(billion US$)'], color='g')
ax2.plot(mega_cities['Metropolitian Area/City'], mega_cities['Metropolitian Population'], 'b-')

ax1.set_xlabel('City')
ax1.set_ylabel('GDP (billion US$)', color='g')
ax2.set_ylabel('Population', color='b')
plt.title('Comparison of Mega Cities by GDP and Population')
ax1.set_xticklabels(mega_cities['Metropolitian Area/City'], rotation=90, ha='center')
fig.tight_layout()
plt.show()
```
