# GDP per Capita Analysis
This notebook calculates and visualizes the GDP per capita for metropolitan areas/cities.

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'data/cities_by_gdp.csv'
data = pd.read_csv(file_path)

data['Official est. GDP(billion US$)'] = pd.to_numeric(data['Official est. GDP(billion US$)'].str.replace(',', ''))
data['Metropolitian Population'] = pd.to_numeric(data['Metropolitian Population'].str.replace(',', '').str.replace(' ', '').str.strip())

data['GDP per Capita'] = data['Official est. GDP(billion US$)'] / data['Metropolitian Population']

top_10_gdp_per_capita = data.sort_values(by='GDP per Capita', ascending=False).head(10)
bottom_10_gdp_per_capita = data.sort_values(by='GDP per Capita', ascending=True).head(10)

# Plotting the top and bottom 10 cities by GDP per capita
plt.figure(figsize=(12, 8))
plt.bar(top_10_gdp_per_capita['Metropolitian Area/City'], top_10_gdp_per_capita['GDP per Capita'], label='Top 10')
plt.bar(bottom_10_gdp_per_capita['Metropolitian Area/City'], bottom_10_gdp_per_capita['GDP per Capita'], label='Bottom 10', color='orange')
plt.xlabel('City')
plt.ylabel('GDP per Capita (US$)')
plt.title('Top and Bottom 10 Cities by GDP per Capita')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.show()
```
