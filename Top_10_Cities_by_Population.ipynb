# Top 10 Cities by Population
This notebook identifies and visualizes the top 10 metropolitan areas/cities with the highest population.

```python

import pandas as pd
import matplotlib.pyplot as plt

file_path = 'data/cities_by_gdp.csv'
data = pd.read_csv(file_path)

data['Metropolitian Population'] = pd.to_numeric(data['Metropolitian Population'].str.replace(',', '').str.replace('', '').str.strip())

top_10_population = data.sort_values(by='Metropolitian Population', ascending=False).head(10)

# Plotting the top 10 cities by Population
plt.figure(figsize=(12, 8))
plt.bar(top_10_population['Metropolitian Area/City'], top_10_population['Metropolitian Population'])
plt.xlabel('City')
plt.ylabel('Population')
plt.title('Top 10 Cities by Population')
plt.xticks(rotation=45, ha='right')
plt.show()
```
