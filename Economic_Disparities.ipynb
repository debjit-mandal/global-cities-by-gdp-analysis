# Economic Disparities
This notebook analyzes economic disparities by comparing GDP distributions across different countries/regions.

```python

import pandas as pd
import matplotlib.pyplot as plt

file_path = 'data/cities_by_gdp.csv'
data = pd.read_csv(file_path)

data['Official est. GDP(billion US$)'] = pd.to_numeric(data['Official est. GDP(billion US$)'].str.replace(',', ''))

country_data = data.groupby('Country/Region')['Official est. GDP(billion US$)'].describe()

# Plotting GDP disparities using box plots
plt.figure(figsize=(20, 10))
data.boxplot(column='Official est. GDP(billion US$)', by='Country/Region', rot=90)
plt.xlabel('Country/Region')
plt.ylabel('GDP (billion US$)')
plt.title('Economic Disparities by Country/Region')
plt.xticks(rotation=90, ha='center', fontsize=5)
plt.suptitle('')
plt.tight_layout()
plt.show()
```
