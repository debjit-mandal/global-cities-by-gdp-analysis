# Urbanization Impact on GDP
This notebook analyzes the impact of urbanization on GDP by grouping cities based on their population sizes.

```python

import pandas as pd
import matplotlib.pyplot as plt

file_path = 'data/cities_by_gdp.csv'
data = pd.read_csv(file_path)

data['Official est. GDP(billion US$)'] = pd.to_numeric(data['Official est. GDP(billion US$)'].str.replace(',', ''))
data['Metropolitian Population'] = pd.to_numeric(data['Metropolitian Population'].str.replace(',', '').str.replace('â€‚', '').str.strip())

population_groups = [1e5, 1e6, 1e7, 1e8]  # Example thresholds
data['Population Group'] = pd.cut(data['Metropolitian Population'], bins=population_groups, labels=['100k-1M', '1M-10M', '10M-100M'])

# Plotting GDP by population group
plt.figure(figsize=(12, 8))
data.groupby('Population Group')['Official est. GDP(billion US$)'].mean().plot(kind='bar')
plt.xlabel('Population Group')
plt.ylabel('Average GDP (billion US$)')
plt.title('Urbanization Impact on GDP')
plt.show()
```
