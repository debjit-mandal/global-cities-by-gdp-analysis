# GDP Distribution by Region
This notebook analyzes the distribution of GDP within different regions.

```python

import pandas as pd
import matplotlib.pyplot as plt

file_path = 'data/cities_by_gdp.csv'
data = pd.read_csv(file_path)

data['Official est. GDP(billion US$)'] = pd.to_numeric(data['Official est. GDP(billion US$)'].str.replace(',', ''))

# Plotting the GDP distribution
regions = data['Country/Region'].unique()
plt.figure(figsize=(14, 10))

for region in regions:
    region_data = data[data['Country/Region'] == region]
    plt.hist(region_data['Official est. GDP(billion US$)'], bins=20, alpha=0.5, label=region)

plt.xlabel('GDP (billion US$)')
plt.ylabel('Number of Cities')
plt.title('GDP Distribution by Region')
plt.legend()
plt.show()
```
